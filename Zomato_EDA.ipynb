```python
df_country = pd.read_excel('Zomatodataset/Country-Code.xlsx')
df_country.head()
```

```python
final_df = pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head()
```

```python
## To check data types
final_df.dtypes
```

```python
country_names = final_df.Country.value_counts().index
country_val = final_df.Country.value_counts().values
```

```python
# Pie chart Top 3 countries that uses zomato
plt.pie(country_val[:3], labels=country_names[:3],autopct= '%1.2f%%')
```

Observation Zomato max records or transactions are from India after USA and UK

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'}) 
ratings 
```

## Observation
1. When rating is between 4.5 to 4.9 -> Excellent
2. When rating is between 4.0 to 4.4 -> Very Good
3. When rating is between 3.5 to 3.9 -> Good
4. When rating is between 2.5 to 3.4 -> Average
5. When rating is between 1.8 to 2.4 -> Poor
6. When rating is 0 -> Not Rated

```python
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x='Aggregate rating',y='Rating count',data=ratings)
```

```python
sns.barplot(x='Aggregate rating',y='Rating count',hue='Rating color',data=ratings,palette=['Blue','red','orange','yellow','green','green'])
```

## Observation:
1. Not rated is very high.
2. Maximum number of rating between 2.5 to 3.4

```python
# Count plot
sns.countplot(x='Rating color',hue = 'Rating text',legend=False, data = ratings, palette=['Blue','red','orange','yellow','green','green'])
```
